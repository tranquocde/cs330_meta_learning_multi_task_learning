```python
import matplotlib.pyplot as plt

# Loss values
loss_values = [0.6, 0.4, 0.3, 0.2, 0.15, 0.12, 0.1, 0.09]

# Number of epochs or iterations
epochs = range(1, len(loss_values) + 1)

# Create the loss function plot
plt.plot(epochs, loss_values,color='red')

# Customize the plot (optional)
plt.title("Loss Function")
plt.xlabel("Epochs")
plt.ylabel("Loss")

# Display the plot
plt.show()
```
